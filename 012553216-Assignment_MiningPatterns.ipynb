# Create a topic modeling visualization

```python
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
newsgroups = fetch_20newsgroups(categories=['sci.med','sci.space','talk.politics.guns'],remove=('headers', 'footers', 'quotes'))
docs_raw = newsgroups.data
print(len(docs_raw))
```

```python
#Term Frequency 
tf_vectorizer = CountVectorizer(strip_accents = 'unicode',
                                stop_words = 'english',
                                lowercase = True,
                                token_pattern = r'\b[a-zA-Z]{3,}\b',
                                max_df = 0.5, 
                                min_df = 10)
dtm_tf = tf_vectorizer.fit_transform(docs_raw)
print(dtm_tf.shape)
```

```python
# Term Frequency Inverse Doc Frequency (Bag of words)
tfidf_vectorizer = TfidfVectorizer(**tf_vectorizer.get_params())
dtm_tfidf = tfidf_vectorizer.fit_transform(docs_raw)
print(dtm_tfidf.shape)
```

```python
# for TF DTM
lda_tf = LatentDirichletAllocation(n_components=3, random_state=0)
lda_tf.fit(dtm_tf)
# for TFIDF DTM
lda_tfidf = LatentDirichletAllocation(n_components=3, random_state=0)
lda_tfidf.fit(dtm_tfidf)
```

```python
pyLDAvis.sklearn.prepare(lda_tf, dtm_tf, tf_vectorizer)
```

```python
pyLDAvis.sklearn.prepare(lda_tfidf, dtm_tfidf, tfidf_vectorizer)
```

## Additional Parameters

```python
# for TF DTM
lda_tf = LatentDirichletAllocation(n_components=10, random_state=0,learning_method = 'online',learning_decay=0.5)
lda_tf.fit(dtm_tf)
# for TFIDF DTM
lda_tfidf = LatentDirichletAllocation(n_components=10, random_state=0,learning_method = 'online',learning_decay=0.6)
lda_tfidf.fit(dtm_tfidf)
```

```python
pyLDAvis.sklearn.prepare(lda_tf, dtm_tf, tf_vectorizer)
```

```python
pyLDAvis.sklearn.prepare(lda_tfidf, dtm_tfidf, tfidf_vectorizer)
```

## Association rule mining

```python
import pandas as pd
df = pd.read_csv('ratings_small.csv')
df.head(3)
```

```python
#Select only movies with ratings larger than 3
df = df[df['rating']>3]
df.head(3)
```

```python
users = df['userId'].unique()
```

```python
import numpy as np
dataset=[]
for i, row in enumerate(users):
    dataset.append(np.append(row,df[df['userId']==1].movieId.iloc[:].values))
```

```python
dataset
```

# Apriori Algorithm to find Frequent sets

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, fpmax, fpgrowth

te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)

frequent_itemsets = apriori(df, min_support=0.15, use_colnames=True)


frequent_itemsets
```
