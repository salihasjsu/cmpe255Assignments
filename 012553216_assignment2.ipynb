```python
import pandas as pd
import numpy as np
source_df = pd.read_csv('./Energy_Consumption_by_Sector_monthly.csv',skiprows=[1])
source_df.head()
```

```python
# Check if there are any null
source_df.count()
```

```python
#Splitting Month columns to month and year
month_year = source_df['Month'].str.split(' ',expand=True)
source_df['Year'] = month_year[0]
source_df['Month'] = month_year[1]
source_df.head()
```

# Part1 plotting graph

```python
import matplotlib.pyplot as plt
%matplotlib inline
source_df.plot(x='Total Energy Consumed by the Residential Sector', y='Month',kind='scatter',figsize=(10,10))
plt.xlabel('Total Energy Consumed by the Residential Sector (Scattered Plot)')  
plt.ylabel('Year')  
  
# displaying the title 
plt.title("Total Energy Consumed by the Residential Sector/Month")
plt.show()
```

```python
month_df = source_df[['Month','Total Energy Consumed by the Residential Sector']]
month_df.groupby('Month')['Total Energy Consumed by the Residential Sector'].sum().plot(kind='line',figsize=(10,10))
plt.xlabel('Month')  
plt.ylabel('Total Energy Consumed by the Residential Sector')  
  
# displaying the title 
plt.title("Total Energy Consumed by the Residential Sector/Month (AggregatebyMonth)")

plt.show()
```

# Part2 aggregate year and plot

```python
new_df = source_df[['Year','Total Energy Consumed by the Residential Sector']]
```

```python
new_df.groupby(new_df['Year'])['Total Energy Consumed by the Residential Sector'].sum().plot(kind='line',y='Total Energy Consumed by the Residential Sector', x='Year',figsize=(10,10))
plt.xlabel('Year')  
plt.ylabel('Total Energy Consumed by the Residential Sector')  
  
# displaying the title 
plt.title("Aggregating Year (sum)") 
plt.show()
```

```python
#new_df.groupby(['Year','Total Energy Consumed by the Residential Sector'])['NEWS_SENTIMENT_DAILY_AVG'].sum().plot(kind='line')
new_df.groupby('Year')['Total Energy Consumed by the Residential Sector'].plot(kind='line',y='Total Energy Consumed by the Residential Sector', x='Year',figsize=(10,10))
plt.xlabel('Year')  
plt.ylabel('Total Energy Consumed by the Residential Sector')  
  
# displaying the title 
plt.title("Aggregating Year") 
plt.xticks(np.arange(0, 600, 50))
plt.show()
```

```python
import ssl

ssl._create_default_https_context = ssl._create_unverified_context
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'
df_wine = pd.read_csv(url,names=['Target','Alcohol','Malic acid','Ash','Alcalinity of Ash','Magnesium','Total phenols',
                              'Flavanoids','Nonflavanoid Phenols','Proanthocyanins','Colour Intensity',
                              'Hue','OD280/OD315 of diluted wines','Proline'])
df_wine.head()
```

# Part 2- b

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

features = ['Alcohol','Malic acid','Ash','Alcalinity of Ash','Magnesium','Total phenols',
                              'Flavanoids','Nonflavanoid Phenols','Proanthocyanins','Colour Intensity',
                              'Hue','OD280/OD315 of diluted wines','Proline']
# spearting out the features
x = df_wine.loc[:, features].values
y = df_wine.loc[:,['Target']].values
x = scaler.fit_transform(x)
result_df = pd.DataFrame(x,columns=['Alcohol','Malic acid','Ash','Alcalinity of Ash','Magnesium','Total phenols',
                              'Flavanoids','Nonflavanoid Phenols','Proanthocyanins','Colour Intensity',
                              'Hue','OD280/OD315 of diluted wines','Proline'])
result_df.head()
```

# Part 2- c

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

features = ['Alcohol','Malic acid','Ash','Alcalinity of Ash','Magnesium','Total phenols',
                              'Flavanoids','Nonflavanoid Phenols','Proanthocyanins','Colour Intensity',
                              'Hue','OD280/OD315 of diluted wines','Proline']
# spearting out the features
x = df_wine.loc[:, features].values
y = df_wine.loc[:,['Target']].values
x = scaler.fit_transform(x)
result_df = pd.DataFrame(x,columns=['Alcohol','Malic acid','Ash','Alcalinity of Ash','Magnesium','Total phenols',
                              'Flavanoids','Nonflavanoid Phenols','Proanthocyanins','Colour Intensity',
                              'Hue','OD280/OD315 of diluted wines','Proline'])
result_df.head()
```

# Part 2- d 
### With Variance 90%

```python
from sklearn.decomposition import PCA
pca = PCA(.90)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             )
finalDf = pd.concat([principalDf, df_wine[['Target']]], axis = 1)
finalDf.head()
```

### With Variance 95%

```python
pca = PCA(.95)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             )
finalDf = pd.concat([principalDf, df_wine[['Target']]], axis = 1)
finalDf.head()
```

# Findings
As we can see with the above result. With 90% variance, the dimensionality is reduced from 13 to 8 features. Wereas if you increase the variace rate, the dimentionality reduced from 13 features to 10 features.
