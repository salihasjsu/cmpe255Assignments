```python
print(__doc__)

# Author: Phil Roth <mr.phil.roth@gmail.com>
# License: BSD 3 clause

import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

plt.figure(figsize=(12, 12))

n_samples = 1500
random_state = 170
X, y = make_blobs(n_samples=n_samples, random_state=random_state)

# Incorrect number of clusters
y_pred = KMeans(n_clusters=2, random_state=random_state).fit_predict(X)

plt.subplot(221)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.title("Incorrect Number of Blobs")
```

```python
for val in range(2,10):
    y_pred = KMeans(n_clusters=val, random_state=random_state).fit_predict(X)
    plt.figure(figsize=(12, 12))
    plt.subplot(221)
    plt.scatter(X[:, 0], X[:, 1], c=y_pred)
    plt.title("Number of Blobs with %d clusters"%val )
    plt.show()
```

## Plot the inertia_ (for  values of n_clusters from 1 to 12 ) of the kmeans clustering model KM trained on the data (KM.inertia_) on the y axis, with the values of n_clusters on the x axi

```python
inertia_axis = []
cluster_axis = []
for val in range(1,13):
    model = KMeans(n_clusters=val, random_state=random_state).fit(X)
    inertia_axis.append(model.inertia_)
    cluster_axis.append(val)
plt.figure(figsize=(12, 12))
plt.subplot(221)
plt.plot(cluster_axis,inertia_axis)
plt.xlabel('Inertia')
plt.ylabel('no of clusters')
plt.show()
```

### As expected The value of inertia decreases as the number of clusters increase.

```python
from yellowbrick.cluster import InterclusterDistance
# Instantiate the clustering model and visualizer
model = KMeans(n_clusters = 4)
visualizer = InterclusterDistance(model)

visualizer.fit(X)        # Fit the data to the visualizer
visualizer.show() 
#reference link : https://www.scikit-yb.org/en/latest/api/cluster/icdm.html
```
