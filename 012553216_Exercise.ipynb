```python
import ssl

ssl._create_default_https_context = ssl._create_unverified_context
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'
df_wine = pd.read_csv(url,names=['Alcohol','Malic acid','Ash','Alcalinity of Ash','Magnesium','Total phenols',
                              'Flavanoids','Nonflavanoid Phenols','Proanthocyanins','Colour Intensity',
                              'Hue','OD280/OD315 of diluted wines','Proline'])
df_wine.head()
```

```python
import pandas as pd    
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
df_scaled = pd.DataFrame(ss.fit_transform(df_wine),columns = df_wine.columns)
df_scaled_cov = df_scaled.cov()
df_scaled_cov.style.background_gradient(cmap='rainbow', axis=None)
```

# Correlation Matrix

```python
cor_dfm = df_wine.corr()
cor_dfm
```

```python
#using dataframe covariance matrix
import seaborn as sns 
import matplotlib.pyplot as plt
f, ax = plt.subplots(figsize=(10, 8)) 
sns.heatmap(df_scaled_cov, mask=np.zeros_like(cor_dfm, dtype=np.bool),cmap=sns.diverging_palette(220, 10, as_cmap=True), square=True, ax=ax) 
```

# Covariance Matrix using Numpy and Pandas + Without Standardization

```python
# using Pandas Dataframe cov method

cov_mat = df_wine.cov()
df_wine.cov().style.background_gradient(cmap='rainbow', axis=None)
```

```python
# using numpy cov method
df_np = df_wine.to_numpy()
cov_mat_np = np.cov(df_np,rowvar=0)
print(cov_mat_np.shape)
dataset = pd.DataFrame(cov_mat_np)
dataset.style.background_gradient(cmap='rainbow', axis=None)
```
